```python
import gymnasium
import safe_autonomy_sims.gym

# Create environment using gymnasium.make()
env = gymnasium.make("Docking-v0")

# Basic interaction loop
observation, info = env.reset()
done = False

while not done:
    # Sample random action from action space
    action = env.action_space.sample()

    # Step the environment
    observation, reward, terminated, truncated, info = env.step(action)
    done = terminated or truncated

env.close()
```

```python
# Inspect environment spaces
print(f"Action space: {env.action_space}")
print(f"Observation space: {env.observation_space}")

# Access episode information
print(f"Status: {info['status']}")
print(f"Reward components: {info['reward_components']}")
```

```python
import safe_autonomy_sims.pettingzoo

# Create environment with specified number of agents
env = safe_autonomy_sims.pettingzoo.MultiDockingEnv(num_agents=2)

# Basic parallel interaction loop
observations, infos = env.reset()

while env.agents:
    # Sample actions for all active agents
    actions = {
        agent: env.action_space(agent).sample()
        for agent in env.agents
    }

    # Step all agents simultaneously
    observations, rewards, terminations, truncations, infos = env.step(actions)

env.close()
```

```python
# Access agent information
print(f"Possible agents: {env.possible_agents}")
print(f"Active agents: {env.agents}")

# Get agent-specific spaces
for agent in env.possible_agents:
    print(f"Agent {agent} action space: {env.action_space(agent)}")
    print(f"Agent {agent} observation space: {env.observation_space(agent)}")
```
